```python
df['begin_time'] = pd.to_datetime(df['begin_time'].str.replace(' UTC', ''))
```

```python
df['request_time'] = pd.to_datetime(df['request_time'].str.replace(' UTC', ''))
```

```python
df['drop_time'] = pd.to_datetime(df['drop_time'].str.replace(' UTC', ''))
```

```python
import datetime
from datetime import datetime
df['year'] = df.request_time.map(lambda x: datetime.strftime(x,"%Y"))
```

```python
df['month'] = df.request_time.map(lambda x: datetime.strftime(x,"%b"))
df['weekday'] = df.request_time.map(lambda x: datetime.strftime(x,"%a"))
df['time'] = df.request_time.map(lambda x: datetime.strftime(x,"%H:%M"))
```

```python
df.head()
```

```python
df['distance_km'] = round(df.dist*1.60934,2)
df['amount_km'] = round(df.fare/df.distance_km,2)
```

```python
df['request_lead_time'] = df.begin_time - df.request_time
df['request_lead_time'] = df['request_lead_time'].apply(lambda x: round(x.total_seconds()/60,1))
```

```python
df['trip_duration'] = df.drop_time - df.begin_time
df['trip_duration'] = df['trip_duration'].apply(lambda x: round(x.total_seconds()/60,1))
```

```python
import numpy as np
df.loc[(df.trip_or_order_status == 'CANCELED') | (df.trip_or_order_status == 'DRIVER_CANCELED'),'request_lead_time']=np.nan
df.loc[(df.trip_or_order_status == 'CANCELED') | (df.trip_or_order_status == 'DRIVER_CANCELED'),'amount_km']=np.nan
df.loc[(df.trip_or_order_status == 'CANCELED') | (df.trip_or_order_status == 'DRIVER_CANCELED'),['begin_time','drop_time']]= np.nan
```

```python
df['completed_ride'] = ((df.trip_or_order_status != 'CANCELED') & (df.trip_or_order_status != 'DRIVER_CANCELED')).astype(int)
```
